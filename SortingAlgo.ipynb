```python
# @title Merge sort
"""
Merge sort implemetation

Time Complexity-
1. sort - O(log N)
2. merge - O(N)

Total - O(N log N) for Best, Worst and Average case.
Con ~ This needs an extra space for sorting array

"""

def sort(listToBeSorted,lowerBound,upperBound):
    if lowerBound == upperBound:
        return
    mid = int((lowerBound + upperBound)/2);

    if lowerBound <= upperBound:
        sort(listToBeSorted,lowerBound,mid)
        sort(listToBeSorted,mid+1,upperBound)
        merge(listToBeSorted,lowerBound,mid,upperBound)

def merge(listToBeSorted,lowerBound,mid,upperBound):
    mergedList = []
    # To track the movement of starting index for left list
    index1 = lowerBound
    # To track the movement of stating index of the right list
    index2 = mid + 1
    # To track the current pointer in merged list
    cursor = 0

    # Run this loop unless one of the list is fully iterated
    while index1 <= mid and index2 <= upperBound:
        if listToBeSorted[index1] <= listToBeSorted[index2]:
            mergedList.insert(cursor,listToBeSorted[index1])
            index1 = index1 + 1
            cursor = cursor + 1
        else :
            mergedList.insert(cursor,listToBeSorted[index2])
            index2 = index2 + 1
            cursor = cursor + 1

    while index1 <= mid:
        mergedList.insert(cursor,listToBeSorted[index1])
        index1 = index1 + 1
        cursor = cursor + 1

    while index2 <= upperBound:
        mergedList.insert(cursor,listToBeSorted[index2])
        index2 = index2 + 1
        cursor = cursor + 1

    # Replace the original list with sorted values
    listToBeSorted[lowerBound:upperBound+1] = (mergedList)

def mergeSort(listToBeSorted):
    #print(listToBeSorted)
    if len(listToBeSorted) == 0 :
        return;
    print("List :",listToBeSorted)
    sort(listToBeSorted,0,len(listToBeSorted)-1);
    print("Final List :",listToBeSorted)


elements = [3,5,1,7,6,239,56,99,232,1002,9,13,23,11,2]
mergeSort(elements)
```

```python
# @title Quick sort
"""
Quick Sort Implementation

Time complexity~
1. quickSort - log N
2. partition - N

Worst time complexity - O(N^2)
Average time complexity - O(N log N) {This is when the pivot is in between and left and righ will have same no. of elements}

Cons ~ Worst case is n^2

Idea - 

"""

def partition(elements,low,high) :
    index = low - 1
    pivot = elements[high]
    lIdx = low

    for value in elements[low:high]:
        if value < pivot:
            index = index + 1
            swap = elements[index]
            elements[index] = value
            elements[lIdx] = swap
        lIdx = lIdx + 1

    index = index + 1
    swap = elements[index]
    elements[index] = pivot
    elements[high] = swap

    return index


def quickSort(elements,low,high):
    #print("Elements ->",elements,"| Low ->",low,"| High ->",high)
    if low < high:
        pivot = partition(elements,low,high)
        quickSort (elements,low,pivot - 1)
        quickSort (elements,pivot + 1,high)

# elements = [3,5,1,9,7,6]
elements = [3,5,1,7,6,239,56,99,232,1002,9,13,23,11,2]
print("Input elements ->",elements)
quickSort(elements,0,len(elements) - 1)
print("Sorted elements ->",elements)

```

```python
# @title Bubble sort
"""
Bubble sort implementation

Time complexity - O (N^2)

Idea - Pushing the largest element at the last
"""
def bubbleSort(elements):
    length = len(elements)
    if length == 0:
        return
    lastIdx =  length - 1
    while lastIdx > 0 :
        for index,value in enumerate(elements[:lastIdx + 1]) :
            if index  < length -1:
                if elements[index] > elements[index + 1]:
                    swap = elements[index]
                    elements[index] = elements[index + 1]
                    elements[index + 1] = swap
        lastIdx = lastIdx -1


elements = [3,5,1,7,6,239,56,99,232,1002,9,13,23,11,2]
print("Input elements ->",elements)
bubbleSort(elements)
print("Sorted elements ->",elements)
```

```python
# @title Insertion sort
"""
Insertion sort

Time complexity = O (N^2)

Idea - Divide the list into left and right half, pick elements from right and put on the sorted position on the left (i.e. sorted list)
"""

def insertionSort(elements) :
    i = 1
    while i <= len(elements) - 1:
        element = elements[i]
        j = i -1
        while j >= 0 and element< elements[j] :
            elements[j+1] = elements[j]
            j = j - 1
        i = i + 1
        elements[j + 1] = element

elements = [3,5,1,7,6,239,56,99,232,1002,9,13,23,11,2]

print("Input elements ->",elements)
insertionSort(elements)
print("Sorted elements ->",elements)
```

```python
# @title Counting sort
"""
Counting sort

Time complexity - O (N + K), n = no of elements, k = range
Space complexity - O(K)

Idea - To create an array of the size of range and put the no. of occurance of the elements in the array comparing with the index of the array
"""

def countingSort(elements, r):
    list = [0]*(r+1)
    for value in elements:
        list[value] = list[value] + 1

    insertIndex = 0
    for index,value in enumerate(list):
        if value > 0:
            for i in range(value):
                elements[insertIndex] = index
                insertIndex = insertIndex + 1


elements = [3,5,1,7,6,239,56,99,232,1002,9,13,23,11,2]
print("Input elements ->",elements)
countingSort(elements,1002)
print("Sorted elements ->",elements)
```

```python
# @title Shell sort
"""
Shell sort

Time complexity - 
Space complexity - 

Idea - 

"""

def shellSort(elements):
    noOfElements = len(elements)
    gap = int (noOfElements/2)
    while gap > 1:
        i = 0
        while i + gap < noOfElements:
            if elements[i] > elements[gap + i]:
                swap = elements[i]
                elements[i] = elements[gap + i]
                elements[i + gap] = swap
            if i + 1 == gap:
                i = i + gap
            i = i + 1
        gap = gap - 1

elements = [3,5,1,7,6,239,56,99,232,1002,9,13,23,11,2]
print("Input elements ->",elements)
shellSort(elements)
print("Input elements after shell short ->",elements)
insertionSort(elements)
print("Sorted elements after insertion short->",elements)
```
